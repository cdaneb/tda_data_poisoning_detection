```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.lineplot(data=tda, x="t", y="score")
plt.axhline(3.0, linestyle="--")  # threshold
plt.title("TDA Detection Score Over Time")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=tda, x="t", y="h1_max_persistence")
plt.title("H1 Max Persistence Over Time")
plt.show()
```

```python
poison_start = 800
poison_end = 1200

plt.figure(figsize=(10, 5))
sns.lineplot(data=tda, x="t", y="score")
plt.axvspan(poison_start, poison_end, alpha=0.2)
plt.show()
```
