```python
import pandas as pd
from pathlib import Path

BASE = Path("..") / "outputs" / "clean" / "seed_0"
metrics = pd.read_csv(BASE / "metrics.csv")

print(metrics.columns.tolist())
print(metrics.head())
```

```python
import numpy as np

# pick the first matching accuracy column
acc_candidates = ["test_acc", "test_accuracy", "accuracy", "acc", "val_acc", "val_accuracy"]
acc_col = next((c for c in acc_candidates if c in metrics.columns), None)

if acc_col is None:
    raise KeyError(f"No accuracy column found. Columns are: {metrics.columns.tolist()}")

print("Using accuracy column:", acc_col)
print(metrics[acc_col].describe())
```

```python
tda = pd.read_csv(BASE / "tda_features.csv")
print(tda.columns.tolist())
print("flags:", int(tda["flag"].sum()) if "flag" in tda.columns else "no flag column")
print("score 99%:", float(tda["score"].quantile(0.99)) if "score" in tda.columns else "no score column")
```

```python
acc = metrics["test_accuracy"]

print("Accuracy summary:")
print(acc.describe())

print("\nMean accuracy:", acc.mean())
print("Min accuracy:", acc.min())
print("Max accuracy:", acc.max())
```
